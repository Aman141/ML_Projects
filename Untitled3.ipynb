```python
import pdfplumber as pp
from gtts import gTTS

pdf_text = ''

with pp.open('Rich Dad Poor Dad.pdf') as pdf:
    for page in pdf.pages:
        pdf_text += page.extract_text()
```

```python
import time
tts = gTTS(text=pdf_text, lang='en')
time.sleep(10)
tts.save('rich_dad_poor_dad.mp3')
```
