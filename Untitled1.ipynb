```python
input_shape = (3, 2000, 1)
```

```python
class MyModel(tf.keras.Model):
    def __init__(self, input_shape= (3,2000,1)):
      super().__init__()
      model = Sequential()
      self.conv1 = Conv2D(input_shape=input_shape, filters=25, kernel_size=(1, 10), strides=(1, 1),
                      padding='valid', activation='linear')

      #model.add(Reshape(target_shape=(3, 1991, 25, 1)))
      self.conv2 = Conv3D(filters=25, kernel_size=(3, 1, 25),
                      data_format='channels_last')
      #model.add(BatchNormalization())
      #model.add(Activation(activation='elu'))
      #model.add(Flatten())
      #model.add(Reshape(target_shape=(1991, 25, 1)))
      #model.add(MaxPool2D(pool_size=(3, 1), strides=(3, 1), data_format='channels_last'))
      #model.add(Dropout(0.5))

      # Conv Pool Block 2
      self.conv3 = Conv2D(filters=50, kernel_size=(10, 25))
      #model.add(BatchNormalization())
      #model.add(Activation(activation='elu'))
      #model.add(MaxPool2D(pool_size=(3, 1), strides=(3, 1)))
      #model.add(Flatten())

      #model.add(Reshape(target_shape=(218, 50, 1)))
      #model.add(Dropout(0.5))

      # Conv Pool Block 3
      self.conv4 = Conv2D(filters=100, kernel_size=(10, 50))
      #model.add(BatchNormalization())
      #model.add(Activation(activation='elu'))
      #model.add(MaxPool2D(pool_size=(3, 1), strides=(3, 1)))
      #model.add(Flatten())

      #model.add(Reshape(target_shape=(69, 100, 1)))
      #model.add(Dropout(0.5))

    def call(self,x):
      x = self.conv1(x)
      x = tf.reshape(x, [1,3,1991,25,1])
      x = self.conv2(x)
      x = BatchNormalization()(x)
      x = Activation(activation = 'elu')(x)
      x = Flatten()(x)
      x = tf.reshape(x, [1,1991,25,1])
      x = MaxPool2D((3,1), strides=(3,1))(x)
      


      return x

```
